# One hotencoding with Top categories

```python
counts = car['brand'].value_counts()

# Define threshold
threshold = 100

# Find brands that appear ≤ threshold times
repl = counts[counts <= threshold].index

# Replace rare brands with 'uncommon'
car['brand'] = car['brand'].replace(repl, 'uncommon')

# Apply one-hot encoding
car_encoded = pd.get_dummies(car['brand'])
car_encoded
```
