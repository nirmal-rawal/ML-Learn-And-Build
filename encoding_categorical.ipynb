```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/customer.csv")
```

```python
df.head(5)
```

```python
df=df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
# Splitting dataset
X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:, 0:2],  # Selecting 'review' and 'education' as features
    df.iloc[:, -1].squeeze(),  # Selecting 'purchased' and ensuring it's 1D
    test_size=0.2,
    random_state=42  # Ensuring reproducibility
)

```

```python
X_train
```

```python
y_train
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
X_train=oe.transform(X_train)
```

```python
X_train
```

```python
oe.categories_
```

```python
X_train
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
y_train
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train=le.transform(y_train)
y_test=le.transform(y_test)
```
